```python
import requests
from bs4 import BeautifulSoup

website = 'https://coast.noaa.gov/htdata/CMSP/AISDataHandler/2024/index.html'
response = requests.get(website)
soup = BeautifulSoup(response.content, 'html.parser')
zipfiles = [a['href'] for a in soup.find_all('a') if a['href'].endswith('.zip')]
```

```python
# Read the contents of the zip_list.txt file into a list
with open('zip_list.txt', 'r') as file:
    zip_list_from_file = [line.strip() for line in file]

# Compare the two lists
missing_in_file = [zipfile for zipfile in zipfiles if zipfile not in zip_list_from_file]
extra_in_file = [zipfile for zipfile in zip_list_from_file if zipfile not in zipfiles]

print("Missing in file:", missing_in_file)
print("Extra in file:", extra_in_file)
```

```python
with open('zip_list.txt', 'w') as file:
    for zipfile in zipfiles:
        file.write(f"{zipfile}\n")
```
