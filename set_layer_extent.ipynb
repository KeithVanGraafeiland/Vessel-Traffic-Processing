```python
import os

import arcpy
# from credentials import username, password
username = '<username>'
password = '<password>'

from arcgis.gis import GIS
import json
```

```python
# Connect to the GIS
gis = GIS("https://esri.mapsdevext.arcgis.com", username, password)

# Search for items that start with "US Vessel Traffic"
search_query = 'title:"US Vessel Traffic*" AND type:"Feature Service"'
search_results = gis.content.search(query=search_query, max_items=1000)

# What is the new extent?
item_extent = {
    'extent' : '[[-118.163782, 33.759425], [-118.076990, 33.716828]]'
    }
```

```python
# Filter and print the results
for item in search_results:
    if item.type == "Feature Service":
        print(f"Title: {item.title}, Type: {item.type}, ID: {item.id}")
        item.update(item_properties=item_extent)
        print(str(item) + ' updated')
```
