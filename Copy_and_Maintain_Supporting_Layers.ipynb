```python
import arcgis
from arcgis.gis import GIS
import os
import arcpy
import zipfile
```

```python
## Feature Services to copy and maintain from NOAA and Army Corps of Engineers
## NOAA Feature Services
ais_base_stations_FS = 'https://coast.noaa.gov/arcgis/rest/services/Hosted/AISBaseStations/FeatureServer/0'
anchorages_FS = 'https://coast.noaa.gov/arcgis/rest/services/Hosted/Anchorages/FeatureServer/0'
aids_to_navigation_FS = 'https://coast.noaa.gov/arcgis/rest/services/Hosted/AtoNs/FeatureServer/0'
danger_zones_FS = 'https://coast.noaa.gov/arcgis/rest/services/Hosted/DangerZonesRestrictedAreas/FeatureServer/0'
vessel_routing_measures_FS = 'https://coast.noaa.gov/arcgis/rest/services/Hosted/VesselRoutingMeasures/FeatureServer/0'
pilot_boarding_areas_FS = 'https://coast.noaa.gov/arcgis/rest/services/Hosted/PilotBoardingAreas/FeatureServer/0'

## Army Corps of Engineers Feature Services
channel_area_FS = 'https://services7.arcgis.com/n1YM8pTrFmm7L4hs/ArcGIS/rest/services/National_Channel_Framework/FeatureServer/1'
```

```python
gdb_folder = r'D:\temp'

def copy_feature_service_to_gdb(feature_service_url, gdb_folder, gdb_name, layer_name):
    gdb_path = os.path.join(gdb_folder, gdb_name)

    if not arcpy.Exists(gdb_path):
        arcpy.management.CreateFileGDB(gdb_folder, gdb_name)

    out_fc = os.path.join(gdb_path, layer_name)
    if arcpy.Exists(out_fc):
        arcpy.management.Delete(out_fc)

    arcpy.management.CopyFeatures(feature_service_url, out_fc)
    print(f"Created: {out_fc}")

    zip_output = os.path.join(gdb_folder, layer_name + '.zip')

    with zipfile.ZipFile(zip_output, 'w', zipfile.ZIP_DEFLATED) as zipf:
        for root, dirs, files in os.walk(gdb_path):
            for file in files:
                if not file.endswith('.lock'):
                    file_path = os.path.join(root, file)
                    arcname = os.path.relpath(file_path, os.path.dirname(gdb_path))
                    zipf.write(file_path, arcname)
    print(f"Created zip file: {zip_output}")
```

```python
copy_feature_service_to_gdb(ais_base_stations_FS, gdb_folder, 'NOAA_ais_base_stations.gdb', 'AIS_Base_Stations')
copy_feature_service_to_gdb(anchorages_FS, gdb_folder, 'NOAA_anchorages.gdb', 'Anchorages')
copy_feature_service_to_gdb(aids_to_navigation_FS, gdb_folder, 'NOAA_aids_to_navigation.gdb', 'Aids_to_Navigation')
copy_feature_service_to_gdb(danger_zones_FS, gdb_folder, 'NOAA_danger_zones.gdb', 'Danger_Zones')
copy_feature_service_to_gdb(vessel_routing_measures_FS, gdb_folder, 'NOAA_vessel_routing_measures.gdb', 'Vessel_Routing_Measures')
copy_feature_service_to_gdb(pilot_boarding_areas_FS, gdb_folder, 'NOAA_pilot_boarding_areas.gdb', 'Pilot_Boarding_Areas')
copy_feature_service_to_gdb(channel_area_FS, gdb_folder, 'ACOE_channel_area.gdb', 'Channel_Area')
```
